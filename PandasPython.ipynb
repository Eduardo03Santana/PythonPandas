```python
# Importando biblioteca pandas no Python
import pandas as pd
import numpy as np
```

```python
# Instanciando DF
wine_df = pd.read_csv("winemag-data_first150k.csv")
```

```python
# Exibindo DF
wine_df.head(15)
```

```python
# Listando os Paises
paises = []
for x in wine_df['country']:
    if x not in paises and x is not np.nan: paises.append(x)
paises.sort()

# Listando as Uvas
uvas = []
for x in wine_df['variety']:
    if x not in uvas and x is not np.nan: uvas.append(x)
uvas.sort()
```

Análise pelo país de produção.

```python
# Media da pontuação por pais
country_mean = wine_df['points'].groupby(by=wine_df['country']).mean()
country_mean_df = pd.DataFrame([[pais, country_mean[pais]] for pais in paises], columns=["Pais", "Avaliação Média"])
display(country_mean_df.sort_values(by=["Avaliação Média"],ascending=False))
```

```python
# Media de preço por pais
country_price_mean = wine_df['price'].groupby(by=wine_df['country']).mean()
country_price_mean_df = pd.DataFrame([[pais, country_price_mean[pais]] for pais in paises], columns=["Pais", "Preço Médio"])
display(country_price_mean_df.sort_values(by=["Preço Médio"],ascending=False))
```

```python
# Média de (p)reço e (a)valiação.
pa_mean = country_price_mean_df.merge(country_mean_df,how='inner',on=['Pais'])
display(pa_mean)
```

```python
pa_mean = pa_mean.eval('CustoBeneficio = `Preço Médio` / `Avaliação Média`')
display(pa_mean.sort_values(by=["CustoBeneficio"]))
```

Análise pelo tipo de uva utilizada no vinho.

```python
# Media da pontuação por tipo de uva
grape_mean = wine_df['points'].groupby(by=wine_df['variety']).mean()
grape_mean_df = pd.DataFrame([[uva, grape_mean[uva]] for uva in uvas], columns=["Uva", "Avaliação Média"])
display(grape_mean_df.sort_values(by=["Avaliação Média"],ascending=False))
```

```python
# Media de preço por tipo de uva
grape_price_mean = wine_df['price'].groupby(by=wine_df['variety']).mean()
grape_price_mean_df = pd.DataFrame([[uva, grape_price_mean[uva]] for uva in uvas], columns=["Uva", "Preço Médio"])

grape_price_mean_df = grape_price_mean_df.dropna(subset=["Preço Médio"])

display(grape_price_mean_df.sort_values(by=["Preço Médio"],ascending=False))
```

```python
# (p)reço e (a)valiação médios
pa_mean_grape = grape_price_mean_df.merge(grape_mean_df,how='inner',on=['Uva'])
display(pa_mean_grape)
```

```python
pa_mean_grape["Custo/Beneficio"] = pa_mean_grape['Preço Médio'] / pa_mean_grape['Avaliação Média']
display(pa_mean_grape.sort_values(by=["Custo/Beneficio"]))
```
